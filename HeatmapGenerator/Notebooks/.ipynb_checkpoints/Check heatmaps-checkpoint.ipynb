# Necessary imports

```python
from heatmaps.run_producer_single import produce_heatmaps
from heatmaps.run_producer import load_model
```

```python
import numpy
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.patches as patches
from Custom_functions_and_classes.datasets import INbreastLoader
```

```python
data = INbreastLoader(root="../Sample_data",all_img_location="Images")
bbox, annot, image = data[2]
```

### transforming bbox relative coordinates to their originals

```python
h, w = image.shape[1:]
scale_factor = numpy.array([w, h, w, h], dtype=float)
b_x_min, b_y_min, b_x_max, b_y_max = (bbox * scale_factor).astype(int)
m_x_min, m_y_min, m_x_max, m_y_max = (bbox * scale_factor).astype(int)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

plt.imshow(image[0])
rect = patches.Rectangle((b_x_min, b_y_min,), (b_x_max - b_x_min), (b_y_max - b_y_min), linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
plt.show()
```

### parameters to load the model

```python
parameters = dict(
    device_type="gpu",
    gpu_number=0,

    patch_size=256,

    stride_fixed=70,
    more_patches=5,
    minibatch_size=100,
    seed=0,

    initial_parameters="../models/sample_patch_model.p",
    input_channels=3,
    number_of_classes=4,

    heatmap_type=[0, 1],
)
```

```python
model, device = load_model(parameters)
```

```python
bbox, annot, image = data[2]

heatmap_benign, heatmap_malignant, = produce_heatmaps(image, model, device)
annot
```
