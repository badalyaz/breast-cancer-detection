## Necessary imports

```python
from Custom_functions_and_classes.datasets import DDSMLoader, INbreastLoader
from heatmaps.run_producer import load_model
from heatmaps import run_producer
import random
import matplotlib.pyplot as pyplot
import numpy
from matplotlib.patches import Rectangle
import matplotlib.patches as patches

import torch.nn
import os
```

## Functions

```python
parameters = dict(
    device_type="gpu",
    gpu_number=0,

    patch_size=256,

    stride_fixed=70,
    more_patches=5,
    minibatch_size=100,
    seed=0,

    initial_parameters="/home/server/other_projects/breast_cancer/HeatmapGenerator/models/sample_patch_model.p",
    input_channels=3,
    number_of_classes=4,
    heatmap_type=[0, 1],
    )


def produce_heatmaps(image_array, model, device):

    patches, case = run_producer.sample_patches_single(
        image_array[0],
        parameters=parameters,
    )

    all_prob = run_producer.get_all_prob(
        all_patches=patches,
        minibatch_size=parameters["minibatch_size"],
        model=model,
        device=device,
        parameters=parameters
    )

    heatmap_malignant, _ = run_producer.probabilities_to_heatmap(
        patch_counter=0,
        all_prob=all_prob,
        image_shape=case[0],
        length_stride_list=case[2],
        width_stride_list=case[1],
        patch_size=parameters['patch_size'],
        heatmap_type=parameters['heatmap_type'][0],
    )
    heatmap_benign, patch_counter = run_producer.probabilities_to_heatmap(
        patch_counter=0,
        all_prob=all_prob,
        image_shape=case[0],
        length_stride_list=case[2],
        width_stride_list=case[1],
        patch_size=parameters['patch_size'],
        heatmap_type=parameters['heatmap_type'][1],
    )


    return heatmap_benign, heatmap_malignant


def make_model(patch, stride, device="gpu"):

    
    if device not in ["cpu", 'gpu']:
        raise ValueError(f"expected device to be 'cpu' or 'gpu', but got {device}")
        

    parameters["patch_size"] = patch
    parameters["stride_fixed"] = stride
    parameters["device"] = device

    random.seed(parameters['seed'])
    
    model, device = load_model(parameters)
    
    return model, device

def plot_heatmaps(benign, malignant):
    
    fig, ax = pyplot.subplots(1, 2, figsize = (10, 4))
    ax[0].imshow(benign)
    ax[1].imshow(malignant)

    ax[0].set_title("Benign")
    ax[1].set_title("Malignant")

    pyplot.show()
    
```

### use device='cpu' if don't have gpu

```python
model, device = make_model(256, 70, device="gpu")
```

```python
data = INbreastLoader(root="./Sample_data", all_img_location="Images")
bbox, annot, image = data[0]
```

## Heatmaps for parameters described above
Results with the original parameters

```python
heatmap_benign, heatmap_malignant = produce_heatmaps(image, model, device)
```

```python
plot_heatmaps(heatmap_benign, heatmap_malignant)
```

## Trying with small stride
Small stride results in smooth outputs

```python
model, device = make_model(256, 30) #stride = 20
```

```python
heatmap_benign, heatmap_malignant = produce_heatmaps(image, model, device)
```

```python
plot_heatmaps(heatmap_benign, heatmap_malignant)
```

The small stride helps the heatmap generator to generate a more expressed benign heatmaps, but the negative side is that the heatmaps are blurred

## Trying with big stride
High stride results in very low resolution

```python
model, device = make_model(256, 140) #stride = 140
```

```python
heatmap_benign, heatmap_malignant = produce_heatmaps(image, model, device)
```

```python
plot_heatmaps(heatmap_benign, heatmap_malignant)
```

The big stride just lowers the resolution of heatmaps and there is nothing special about it. the generated features have  very low quality with this parameter.

## Trying with small patch size
Small patch size results in low resolution

```python
model, device = make_model(128, 70) #patch size = 128
```

```python
heatmap_benign, heatmap_malignant = produce_heatmaps(image, model, device)
```

```python
plot_heatmaps(heatmap_benign, heatmap_malignant)
```

As we see, the small patch size with the same stride again empowers the malignant heatmap more. But, as we need to have more expressed benign heatmaps these parameters are not useful for us.

## Trying with big patch size
Big patch size results in very smooth outputs

```python
model, device = make_model(512, 70) #patch size = 512
```

```python
heatmap_benign, heatmap_malignant = produce_heatmaps(image, model, device)
```

```python
plot_heatmaps(heatmap_benign, heatmap_malignant)
```

The bigger is the patch size, better is the Benign heatmap expressed. The big patch size with relatively small stride makes the heatmap blurry. The reason of that is the rescaling of small heatmap to the size of original image.
